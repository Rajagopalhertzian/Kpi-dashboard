```python
import seaborn as sns
import matplotlib.pyplot as plt

# Identify max and min savings and production efficiency
max_saving = df['savings'].max()
min_saving = df['savings'].min()
max_efficiency = df['production_efficiency'].max()
min_efficiency = df['production_efficiency'].min()

# Created a dataframe for max and min values
max_min_df = pd.DataFrame({
    'Metric': ['Max Savings', 'Min Savings', 'Max Efficiency', 'Min Efficiency'],
    'Value': [max_saving, min_saving, max_efficiency, min_efficiency]
})

# Plot max and min values
plt.figure(figsize=(10, 6))

sns.barplot(x='Metric', y='Value', data=max_min_df, palette='viridis')
plt.title('Maximum and Minimum Savings and Production Efficiency')
plt.ylabel('Value')
plt.xlabel('Metric')

plt.show()

# Scatter plot for efficiency vs savings, highlighting max and min
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['production_efficiency'], y=df['savings'], alpha=0.6, color='gray', label='Data Points')

# Highlighting specific points
plt.scatter(max_efficiency, max_saving, color='green', label='Max Efficiency & Saving', s=100)
plt.scatter(min_efficiency, min_saving, color='red', label='Min Efficiency & Saving', s=100)

plt.title('Production Efficiency vs Savings (Highlighting Max and Min)')
plt.xlabel('Production Efficiency (%)')
plt.ylabel('Savings')
plt.legend()
plt.show()
```
